7. Plot their cumulative gross returns and write down what stock had the largest price increase.

```python
sample_gross_returns = sample_prices/sample_prices.shift(1)
```

```python
cum_sample_gross_returns = sample_gross_returns.cumprod()
```

```python
cum_sample_gross_returns.plot(figsize = (12,6), grid = True)
plt.legend(prop = {'size': 9})
plt.ylabel('cumulative gross return')
```

```python
# ETN(Eaton Corporation) had the largest price increase.
```

8. Which stock had the largest average return.

```python
log_gross_returns = np.log(sample_gross_returns)
avg_log_gross_returns = log_gross_returns.mean()
avg_gross_returns = np.exp(avg_log_gross_returns)
avg_gross_returns.sort_values(ascending = False)
```

```python
# ETN(Eaton Corporation) had the largest average return.
```
